**Send a HTTP Request to Fetch the HTML Content of the Page**

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.linkedin.com/jobs/search/?currentJobId=3761416060&keywords=data%20analyst&origin=JOBS_HOME_KEYWORD_AUTOCOMPLETE&refresh=true"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content,'html.parser')
print(soup.prettify())
```

```python
job_titles = [title.text.strip() for title in soup.find_all('h3', class_='base-search-card__title')]

# Extract Company from LinkedIn
company_title=[company.text.strip() for company in soup.find_all('h4',class_='base-search-card__subtitle')]

# Extract Location from LinkedIn
location_title=[location.text.strip() for location in soup.find_all('span',class_='job-search-card__location')]

# Extract Time from LinkedIn
uploaded=[time.text.strip() for time in soup.find_all('time', class_='job-search-card__listdate')]


# Extract Company from LinkedIn
job_link=[link['href'] for link in soup.find_all('a',class_='base-card__full-link')]


# Print the cleaned job titles for verification
print(job_titles)
print(company_title)
print(location_title)
print(uploaded)
print(job_link)
```

```python
import csv
rows=zip(job_titles,company_title,location_title,uploaded,job_link)

#as per the instruction we are making a csv file to store the data extracted from the linkedin
rows=zip(job_titles,company_title,location_title,uploaded,job_link)
# Open the CSV file in write mode
with open('C:\\Users\\ADMIN\\Desktop\\file manager\\WebScrappedData.csv', 'w', newline='', encoding='utf-8') as file:
    # Create a CSV writer object
    writer = csv.writer(file)

    # Write a header if needed
    writer.writerow(['Job Title','company_titles','location_title','uploaded','job_link'])
    # Write the cleaned job titles to the CSV file
    for row in rows:
        writer.writerow(row)
```
